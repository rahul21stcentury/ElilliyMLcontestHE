```python
import os
import pandas as pd
import numpy as np
np.random.seed(1337)
```

```python
cwd = os.getcwd()
train_path=cwd+'\\train_txt'
test_path=cwd+'\\test_txt'
resume_path=cwd+'\\dataset'
```

```python
obj=open('Myfeatures.txt')
obj=obj.read().split(',')
objects=obj
```

```python
objects
```

```python
def traintable(path):
    files=os.listdir(path)

    
    table=pd.DataFrame(columns=[i.lower() for i in objects])

    for i in files[:]:
        f=open(train_path+'\\'+i)
        lines=f.read().lower()  
        #lines
        df=pd.DataFrame({j.lower() : [0] for j in objects})
        df['CandidateID']=i[:-4]
        for k in objects:
            if(k.lower() in lines):
                df[k.lower()]=1
        table=pd.concat([table,df])



    lines



    #files

    train_data=pd.read_csv('dataset\\train.csv')
    train_data.reset_index(drop=True, inplace=True)
    train_data=train_data.sort_values(['CandidateID'])

    train=pd.DataFrame()
    train['CandidateID']=train_data['CandidateID']
    train['Match Percentage']=train_data['Match Percentage']
    train.reset_index(drop=True, inplace=True)

    train.shape

    train

    table.reset_index(drop=True, inplace=True)

    table

    return pd.merge(table,train,on='CandidateID')
```
